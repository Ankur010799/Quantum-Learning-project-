# MODEL BUILDING AND TESTING

```python
y = f["Fraud_Risk"]
X = f.drop(["Fraud_Risk"],axis=1)
```

```python
# Create Train & Test Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
y_test.value_counts()
```

```python
# Running logistic regression model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix 

LR = LogisticRegression().fit(X_train, y_train)
lr_pred = LR.predict(X_test)
print("Train accuracy=",LR.score(X_train, y_train))
# Print the prediction accuracy
print("Test accuracy=",accuracy_score(y_test, lr_pred))
```

```python
# Create the Confusion matrix
print(confusion_matrix(y_test,lr_pred))  
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=1000 , oob_score = True, n_jobs = -1,
                                  random_state =50, max_features = "auto",
                                  max_leaf_nodes = 30)
rfc.fit(X_train, y_train)

print("Train accuracy=",rfc.score(X_train, y_train))
# Make predictions
rfc_pred = rfc.predict(X_test)
print ("Test accuracy=",accuracy_score(y_test, rfc_pred))
```

```python
# Create the Confusion matrix
print(confusion_matrix(y_test,rfc_pred))  
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear').fit(X_train,y_train)

print("Train accuracy=",svm.score(X_train, y_train))
svm_pred = svm.predict(X_test)
print("Test accuracy=",accuracy_score(y_test, svm_pred))
```

```python
# Create the Confusion matrix
print(confusion_matrix(y_test,svm_pred))  
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()
xgb.fit(X_train, y_train)

print("Train accuracy=",xgb.score(X_train, y_train))
xgb_pred = xgb.predict(X_test)
print("Test accuracy=",accuracy_score(y_test, xgb_pred))
```

```python
# Create the Confusion matrix
print(confusion_matrix(y_test,xgb_pred))  
```
